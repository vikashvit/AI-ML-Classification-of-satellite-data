# Step 1 :- read the Image file "3.gif" and convert it into 2D matrix.

```python
import numpy as np
from PIL import Image, ImageSequence
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.cluster import KMeans
img = Image.open('3.gif')
frames = np.array([np.array(frame.copy().convert('RGB').getdata(),dtype=np.uint8).reshape(frame.size[1],frame.size[0],3) for frame in ImageSequence.Iterator(img)])
newarr = frames[0].transpose(2,0,1).reshape(3,-1)
observations_All = newarr[0].reshape(512,512);
print(observations_All)
plt.figure(figsize = (15,8))
plt.imshow(observations_All)
```

# Image "2.gif"  Use KMeans with as n_clusters = 30 and plot the graph using Euclidian Distance

```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import ndimage
from sklearn import cluster


x, y = observations_All.shape
image_2d = observations_All.reshape(x*y, 1)
image_2d.shape


kmeans_cluster = cluster.KMeans(n_clusters = 30, init = 'k-means++', max_iter = 500, n_init = 10, random_state = 0)
kmeans_cluster.fit(observations_All)
cluster_centers = kmeans_cluster.cluster_centers_
cluster_labels = kmeans_cluster.labels_

plt.figure(figsize = (15,8))
plt.imshow(cluster_centers[cluster_labels].reshape(x, y))
```
